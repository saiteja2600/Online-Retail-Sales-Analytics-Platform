```sql
SELECT 
    ps.product_type,
    ROUND(SUM(ps.total_net_sales), 2) AS total_sales,
    ROUND(
        (SUM(ps.total_net_sales) / SUM(SUM(ps.total_net_sales)) OVER ()) * 100,
        2
    ) AS revenue_percentage,
    concat(
        ROUND(
            (SUM(ps.total_net_sales) / SUM(SUM(ps.total_net_sales)) OVER ()) * 100,
            2
        ),
        "%"
    ) AS revenue
FROM products_sales ps
GROUP BY ps.product_type
ORDER BY total_sales DESC
LIMIT 5;
```
