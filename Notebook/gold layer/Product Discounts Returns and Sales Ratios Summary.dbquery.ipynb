```sql
SELECT 
    ps.product_type,
    ROUND(SUM(ps.discounts), 2) AS total_discounts,
    ROUND(SUM(ps.returns), 2) AS total_returns,
    ROUND((SUM(ps.discounts) / SUM(ps.gross_sales)) * 100, 2) AS discount_percentage,
    ROUND((SUM(ps.returns) / SUM(ps.gross_sales)) * 100, 2) AS return_percentage
FROM products_sales ps
GROUP BY ps.product_type
ORDER BY total_discounts LIMIT 10;
```
