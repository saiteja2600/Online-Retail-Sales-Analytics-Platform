```sql
SELECT 
    ps.product_type,
    ROUND(SUM(ps.net_quantity), 2) AS total_quantity_sold,
    ROUND(SUM(ps.gross_sales), 2) AS total_gross_sales,
        ROUND((ABS(SUM(ps.discounts)) / SUM(ps.gross_sales)) * 100,2) AS discount,
    concat(ROUND((ABS(SUM(ps.discounts)) / SUM(ps.gross_sales)) * 100,2) ,"%")AS discount_percentage,
ROUND((ABS(SUM(ps.returns)) / SUM(ps.gross_sales)) * 100,2) AS returns,
concat(ROUND((ABS(SUM(ps.returns)) / SUM(ps.gross_sales)) * 100,2),"%") AS return_percentage,
    ROUND(SUM(ps.total_net_sales), 2) AS total_net_sales
FROM products_sales ps
GROUP BY ps.product_type
ORDER BY total_net_sales DESC
LIMIT 5;
```
