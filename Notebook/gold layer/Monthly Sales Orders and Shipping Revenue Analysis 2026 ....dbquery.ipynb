```sql
SELECT 
    ms.year,
    ms.month,
    SUM(ms.total_orders) AS total_orders,
    ROUND(SUM(ms.shipping),2) AS shipping_revenue,
    ROUND((SUM(ms.shipping) / SUM(ms.total_sales)) * 100,2) AS shipping_percentage,
    ROUND(SUM(ms.total_sales),2) AS total_sales
FROM monthlysales ms
GROUP BY ms.year, ms.month
ORDER BY total_orders DESC


```
