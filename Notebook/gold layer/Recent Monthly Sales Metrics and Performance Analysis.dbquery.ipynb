```sql
SELECT 
    ms.year,
    ms.month,
    ROUND(SUM(ms.total_orders), 2) AS total_orders,
    ROUND(SUM(ms.gross_sales), 2) AS gross_sales,
        ROUND((ABS(SUM(ms.discounts)) / SUM(ms.gross_sales)) * 100, 2) AS discount,

    CONCAT(ROUND((ABS(SUM(ms.discounts)) / SUM(ms.gross_sales)) * 100, 2), "%") AS discount_percentage,
        ROUND((ABS(SUM(ms.returns)) / SUM(ms.gross_sales)) * 100, 2) AS returns,

    concat(ROUND((ABS(SUM(ms.returns)) / SUM(ms.gross_sales)) * 100, 2),"%") AS return_percentage,
    ROUND(SUM(ms.net_sales), 2) AS net_sales,
    ROUND(SUM(ms.shipping), 2) AS shipping,
    ROUND(SUM(ms.total_sales), 2) AS total_sales
FROM monthlysales ms
GROUP BY ms.year, ms.month
ORDER BY total_orders DESC
LIMIT 10;
```
